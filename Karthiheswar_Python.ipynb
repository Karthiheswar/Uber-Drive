## Q25. Make the following ranges of miles ["<5","5-10","10-20","20-35",">35"]. Keep these in the horizontal axis and count of trips in the vertical axis. Plot a bar plot. (2 points)

```python
mildist=df["MILES*"]
milerange_list=["<=5","5-10","10-15","15-20",">20"]
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        plt.text(rect.get_x()+rect.get_width()/2., 1.03*height, '%s' % int(height))
miledict=dict()
for item in milerange_list:
    miledict[item]=0
for mile in mildist.values:
    if mile<=5:
        miledict["<=5"]+=1
    elif mile<=10:
        miledict["5-10"]+=1
    elif mile<=15:
        miledict["10-15"]+=1
    elif mile<=20:
        miledict["15-20"]+=1
    else:
        miledict[">20"]+=1
mildist=pd.Series(miledict)
mildist.sort_values(inplace=True,ascending=False)
print("Miles:\n",mildist)
rects=plot.bar(range(1,len(mildist.index)+1),mildist.values)
plot.title("Miles")
plot.xlabel("Miles")
plot.ylabel("Quantity")
plot.xticks(range(1,len(mildist.index)+1),mildist.index)
plot.grid()
```
